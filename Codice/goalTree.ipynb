```python
from langchain_ollama import ChatOllama

model = ChatOllama(model="llama3.2")
```

```python
from bigtree import *

root = Node("Goal")
budget = Node("Budget", type="Epistemic", value=0, parent=root)
localita = Node("Località", type="Epistemic", value=0, parent=root)
durata = Node("Durata", type="Epistemic", value=0, parent=root)
sub1 = Node("Subgoal 1", type="Epistemic", value=0, parent=budget)
sub2 = Node("Subgoal 2", type="Epistemic", value=0, parent=localita)
sub3 = Node("Subgoal 3", type="Epistemic", value=0, parent=durata)
```

```python
def ratingCalculator(list, budget, localita, durata):
    ms = [m.lower() for m in list]
    
    msc = "".join(ms)
    
    if "budget" in msc:
        budget.value += 1
        
    if "località" in msc:
        localita.value += 1
    
    if "durata" in msc:
        durata.value += 1
```

```python
from langchain_core.messages import HumanMessage, SystemMessage
from langchain_core.output_parsers import StrOutputParser

parser = StrOutputParser()

msg = input("User Prompt")

messages = [
    SystemMessage(content="Analizza attentamente l'user prompt. Scrivi 'budget' se l'utente fa riferimento ad una cifra, un range di spesa, o utilizza parole legate al costo come: 'costo', 'prezzo', 'economico', 'caro', 'spendere', 'lusso', ecc. Scrivi 'località' se l'utente menziona un luogo specifico, come nomi di città, paesi, regioni o luoghi geografici come: 'mare', 'montagna', 'isola', ecc. Scrivi 'durata' solo se l'utente indica esplicitamente una quantità di tempo (giorni, settimane, mesi) o usa termini legati alla durata (es. 'breve', 'lungo', 'settimana', 'weekend'). Non scrivere 'durata' se non viene menzionata. Restituisci esclusivamente le parole chiave rilevanti (budget, località, durata), separate da virgole. Se nessuna categoria è menzionata, restituisci 'null'. Non aggiungere altre parole o spiegazioni."),
    HumanMessage(content=msg),
]

chain = model | parser

count = 0

for i in range(100):
    c = chain.invoke(messages)
    print(c)
    print("\n")
    ratingCalculator(c, budget, localita, durata)
    
    if budget.value == 1 and localita.value == 1 and durata.value == 0:
        count += 1
    
    budget.value = 0
    localita.value = 0
    durata.value = 0 
    
print(count)
```
